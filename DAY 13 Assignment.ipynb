__6. Filter for different condition - come up with insights from the dataset , basic plot , distribution plots__

```python
diff = cars['Class'].value_counts()
diff
```

```python
fraud = cars['Class'].str.contains('1', na=False)
cars[fraud]
```

```python
not_fraud = cars['Class'].str.contains('0', na=False)
cars[not_fraud]
```

```python
x = np.array(cars['Time'])
y = np.array(cars['Amount'])

print(plt.plot(x,y))

print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```

```python
m = np.array(cars.head(10)['Time'])
n = np.array(cars.head(10)['Amount'])

print(plt.bar(m,n))

print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```

```python
x1 = np.array(cars.head(15)['Time'])

print(plt.hist(x1))

print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```
